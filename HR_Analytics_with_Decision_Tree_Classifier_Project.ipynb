### EDA (Exploratory Data Analysis)

```python
# Write a function that makes plot for feature and attrition as hue

def make_plot(feature_name):
    plt.figure(figsize = (15,6))
    a = sns.countplot(x = feature_name ,data= df , hue = 'Attrition')
    plt.xlabel(feature_name)
    plt.title('{} vs Attrition'.format(feature_name))
    plt.legend()

```

```python
make_plot('BusinessTravel')
```

```python
### People who travel frequently are more tend to leave.We are looking at total numbers and ratio of attrition as well.
### While 3000 people are travelling rarely Out of which 2500 are planning to stay back and 500 are  tend to leave. It has a rough ratio of 5:1
### While 750 people travel frequently out of which 550 are planning to stay back and 200 are tend to leave .It has a rough ratio of 3:1.
### Looking at this definetely we need to put our focus and attention on retaining the employees that travel more frequently
### as they more tend to leave the Job.We can increase their salary ,give them an easy project to work on ,or reduce their working hours or change their manager.
```

```python
## Employees working for HR department are more tend to leave
```

```python
## There is no relation of attrition with the education of an employee.
## Education does not appear to have significant effect on attrition.
## As population for a particular category is increasing attrition is increasing.And as population is decreasing attrition is decresing as well.
## There is no particular pattern about attrition we can point out on the basis of Education column.
```

```python
## One hot encoding
one_hot_var = ['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']

for i in one_hot_var:
    val = pd.get_dummies(df[i],prefix = i,prefix_sep = '_',drop_first = True)
    df = df.join(val)
    df.drop(i,axis = 1, inplace = True)
```

```python
df
```

```python
## Encoding values for attrition and Gender columns

df.Attrition =df.Attrition.apply(lambda x: 1 if x == 'Yes' else 0)
df.Gender =df.Gender.apply(lambda x: 1 if x == 'Male' else 0)
```

```python
x = df.drop('Attrition', axis = 1)
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25,random_state = 42)
```

```python
x_train.drop('Department_Sales',axis = 1 ,inplace = True)
x_test.drop('Department_Sales',axis = 1, inplace = True)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()

dtc.fit(x_train,y_train)
```

```python
y_pred = dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```
